```python
# divide the string "text" into a list, by splitting it on every newline ("\n")
text = text.split('\n')
```

```python
# set up an empty list
LU = []

# loop through e
for utterance in text:
    words = utterance.split()
    print(words)
    num_words = len(words)
    LU.append(num_words)
```
