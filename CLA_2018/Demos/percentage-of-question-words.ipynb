```python
#find the total number of words from the Brown corpus and count total number of questions words

#import functions for working with the file system
import os
import glob

#import list of punctuation marks so we can remove these later
from string import punctuation as punct

# point to folder with home-made functions
fun_folder = '/Users/ethan/Documents/GitHub/CLA/Functions'
os.chdir(fun_folder)

# import home-made function for getting only mother or child's speech
import getmotchi

# direct Python to cycle through the folders with the different children's names
children = ['Adam', 'Sarah', 'Eve']
for child in children:
    pathin = '/Users/ethan/Documents/GitHub/CLA/Data/Brown/'+child
    
    # change to the path for each child's folder in turn
    os.chdir(pathin)
    
    # set up an empty list to hold all the mother's words
    words = []
    
    # set up counters for counting the question words
    what = 0
    where = 0
    why = 0
    how = 0
    do = 0
    
    # use the getmotchi function to get only the mother's speech from the trancscript
    for file in glob.glob('*.cha'):
            text = getmotchi.getmot(file)
            for i, val in enumerate(text):
                
                # remove the part of the transcript that is just "*MOT:"
                val = val[5:]
                
                # put the words in the list "words"
                words.append(val)
                
    # make the list "words" into a string, separted by a space
    words = ' '.join(words)
    
    # remove punctuation marks
    words = ''.join(x for x in words if x not in punct)
    
    # put all words in lower case to avoid confusion
    words = words.lower()
    
    # split words into a list, so we can loop through them
    words = words.split()
    
    # count these question words
    for i in words:
        if i == 'what':
            what = what + 1
        if i == 'where':
            where = where +1
        if i == 'why':
            why = why + 1
        if i == 'how':
            how = how + 1
        if i == 'do':
            do = do +1
    
    # calculate total number of words for each child, and percentage of total that is each of the question words
    words = len(words)
    print('Total words for ' + child + ': ' + str(words))
    what = what/words*100
    where = where/words*100
    why = why/words*100
    how = how/words*100
    do = do/words*100
    
    # display the answers
    print('% what: ' + str(what))
    print('% where: ' + str(where))
    print('% why: ' + str(why))
    print('% how: ' + str(how))
    print('% do: ' + str(do))
    print(' ')

            
```
